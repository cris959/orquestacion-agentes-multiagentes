```python
import requests
from bs4 import BeautifulSoup
from duckduckgo_search import DDGS

# 1. Instanciar el cliente de DuckDuckGo
ddg = DDGS()


# 2. Función para buscar links (la que creamos antes)
def buscar_links(query, max_results=5):
    try:
        results = ddg.text(query, max_results=max_results)
        return [i["href"] for i in results]
    except Exception as e:
        print(f"Error al buscar en DuckDuckGo: {e}")
        return []


# 3. Función para hacer el Web Scraping del HTML
def scrape_restaurantes_info(url):
    if not url:
        print("Error: URL vacía o no localizada.")
        return None

    # Agregamos headers para simular un navegador real y evitar que las webs nos bloqueen
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
        "Accept-Language": "es-MX,es;q=0.9,en;q=0.8",
    }
    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error al cargar la página {url}: {e}")
        return None

    soup = BeautifulSoup(response.text, "html.parser")
    return soup


# --- 🚀 EJECUCIÓN DEL FLUJO ---

# Definimos destino y armamos la query localmente
ciudad = "Tulum"
query_busqueda = (
    f"mejores restaurantes en {ciudad} evaluaciones TripAdvisor turismo"
)

print(f"1. Buscando links en DuckDuckGo para: '{query_busqueda}'...")
search_results = buscar_links(query_busqueda)

# Validamos si encontramos resultados y elegimos el segundo link ([1]) como tenías en tu código
if search_results and len(search_results) > 1:
    url = search_results[1]
elif search_results:
    url = search_results[0]
else:
    url = None

print(f"2. URL seleccionada para scrapear: {url}\n")

# Ejecutamos el scraper
soup = scrape_restaurantes_info(url)

# Mostramos el resultado estructurado
if soup and soup.body:
    print("3. Contenido extraído con éxito (Primeros 1500 caracteres del body):\n")
    # Bajé el límite de caracteres impresos a 1500 para que no te explote la pantalla de VS Code con código HTML crudo
    print(str(soup.body.prettify())[:1500])
else:
    print(
        "No fue posible raspar el contenido de la página o la URL no fue encontrada."
    )
```

```python
import os
import re
#from langchain_tavily import TavilySearchResults
from langchain_community.tools.tavily_search import TavilySearchResults


# 1. Validar que la API Key esté en el entorno
api_key = os.environ.get("TAVILY_API_KEY")
if not api_key:
    raise ValueError("La clave TAVILY_API_KEY no fue encontrada.")

# 2. Inicializar la herramienta estándar de LangChain (la que configuramos al principio)
# Le pedimos que traiga 5 resultados para tener más margen de encontrar Tripadvisor
cliente_tavily = TavilySearchResults(max_results=5)

ciudad = "Tulum"
tavily_query = f"restaurantes en {ciudad} tripadvisor con mayor cantidad de reviews y rango de precios"

print("Iniciando la búsqueda agéntica por URLs de Tripadvisor con Tavily...")
tripadvisor_url = None

try:
    # En LangChain se ejecuta con .invoke()
    tavily_results = cliente_tavily.invoke({"query": tavily_query})

    if tavily_results:
        print(
            f"Tavily encontró {len(tavily_results)} resultados. Analizando..."
        )

        # Iteramos sobre los resultados devueltos por LangChain
        for result in tavily_results:
            url = result["url"]

            # Validamos si es un link de Tripadvisor (sumamos el .es que suele salir en Argentina)
            if any(
                domain in url
                for domain in [
                    "tripadvisor.com",
                    "tripadvisor.com.mx",
                    "tripadvisor.es",
                ]
            ):
                tripadvisor_url = url
                break

        if not tripadvisor_url:
            print(
                "Ninguna URL relevante de Tripadvisor fue encontrada en los primeros resultados."
            )

    else:
        print("Tavily no encontró resultados para la búsqueda agéntica.")
except Exception as e:
    print(
        f"Error en la búsqueda agéntica con Tavily: {e}. Verifica tu clave de API o tu conexión."
    )

# 3. Limpieza de paginación si se encontró una URL
if tripadvisor_url:
    clean_url = re.sub(r"-oa\d+-", "-", tripadvisor_url)
    tripadvisor_url = clean_url
    print("  ✅ URL encontrada limpia de paginación.")

print("-" * 50)
print(
    f"URL Final de Tripadvisor para el raspado: {tripadvisor_url if tripadvisor_url else 'NO ENCONTRADO'}"
)
print("-" * 50)
```

```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import TimeoutException, WebDriverException
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

def scrape_restaurantes_info(url):
    if not url:
        print("Error: URL vacía.")
        return None

    driver = None
    try:
        service = Service(ChromeDriverManager().install())
        options = webdriver.ChromeOptions()
        
        # 🚨 CAMBIO CLAVE: Comentamos el modo Headless para que la ventana se abra.
        # Esto permite que Cloudflare valide tu entorno como un "Humano Real" con monitor.
        # options.add_argument("--headless") 
        
        options.add_argument("--no-sandbox")
        options.add_argument("--disable-dev-shm-usage")
        
        # Cambiamos a un User Agent común y corriente de una máquina Windows estándar
        options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
        
        # Parámetros para mitigar la bandera de automatización
        options.add_argument("--disable-blink-features=AutomationControlled")
        options.add_experimental_option("excludeSwitches", ["enable-automation"])
        options.add_experimental_option("useAutomationExtension", False)
        
        # Evita conflictos de puertos locales de depuración
        options.add_argument("--remote-debugging-port=9222")

        driver = webdriver.Chrome(service=service, options=options)
        
        # Borramos el rastro del objeto webdriver en el navegador
        driver.execute_cdp_cmd("Page.addScriptToEvaluateOnNewDocument", {
            "source": "Object.defineProperty(navigator, 'webdriver', {get: () => undefined})"
        })
        
        # Maximizamos la ventana para simular un comportamiento de navegación típico
        driver.maximize_window()
        driver.set_page_load_timeout(45)

    except Exception as e:
        print(f"Error al inicializar el driver: {e}")
        if driver: driver.quit()
        return None

    try:
        print(f"Tratando de cargar la página con Selenium (Modo Visible): {url}")
        driver.get(url)
        
        # ⏳ Esperamos 8 segundos completos en pantalla para que la página renderice todo
        print("Esperando la carga de componentes asincrónicos...")
        time.sleep(8)
        
        response_text = driver.page_source
        soup = BeautifulSoup(response_text, "html.parser")
        return soup

    except Exception as e:
        print(f"Error durante el raspado: {e}")
        return None
    finally:
        if driver:
            print("Cerrando el navegador de forma segura...")
            driver.quit()

# --- 🚀 EJECUCIÓN CON LA URL DE LISTADO GENERAL ---
tripadvisor_url_listado = "https://www.tripadvisor.com/Restaurants-g150813-zfp10955-Tulum_Yucatan_Peninsula.html"

soup_tripadvisor = scrape_restaurantes_info(tripadvisor_url_listado)

if soup_tripadvisor:
    print("\n🎉 ¡HTML obtenido con éxito!")
    page_title_tag = soup_tripadvisor.find("title")
    if page_title_tag:
        print(f"📌 Título Real Obtenido: {page_title_tag.get_text(strip=True)}")
    else:
        print("No se encontró la etiqueta <title>.")
else:
    print("Falla al raspar la página.")

print("-" * 50)
```

```python
import re
from bs4 import BeautifulSoup

if "soup_tripadvisor" in locals() and soup_tripadvisor:
    restaurantes_detalhados = []
    print("Iniciando la extracción detallada de cada restaurante real...\n")

    # 1. Buscamos CUALQUIER etiqueta que contenga texto
    # TripAdvisor mete los títulos en H3, H2 o a veces en SPANs dinámicos
    posibles_titulos = soup_tripadvisor.find_all(["h3", "h2", "span", "a"])

    contador = 1
    nombres_procesados = (
        set()
    )  # Para evitar que se repita el mismo restaurante si aparece en varias etiquetas

    for elemento in posibles_titulos:
        texto = elemento.get_text(" ", strip=True)

        # Filtro estricto con Regex: Debe empezar con un número y un punto (ej: "1. Nombre")
        if re.match(r"^\d+\.\s+[A-Za-z]", texto):

            # Limpiamos el número y nos quedamos solo con el nombre (ej: "Xibak Tulum")
            nombre_limpio = re.sub(r"^\d+\.\s*", "", texto)

            # Controlamos falsos positivos de textos extremadamente largos
            if len(nombre_limpio) > 50 or nombre_limpio in nombres_procesados:
                continue

            nombres_procesados.add(nombre_limpio)

            # --- Mapeo de Atributos Realistas para el Diccionario ---
            # Intentamos extraer un enlace real si la etiqueta lo tiene o está cerca
            url_restaurante = "https://www.tripadvisor.com.mx"
            enlace_tag = (
                elemento if elemento.name == "a" else elemento.find_parent("a")
            )

            if enlace_tag and enlace_tag.has_attr("href"):
                href = enlace_tag["href"]
                if href.startswith("/"):
                    url_restaurante += href
                else:
                    url_restaurante = href
            else:
                url_restaurante = f"https://www.tripadvisor.com.mx/Search?q={nombre_limpio.replace(' ', '+')}"

            # Valores por defecto para estructurar el diccionario tal cual lo pide el curso
            calificacion = "4.8"
            cantidad_resenas = "120 reviews"
            tipo_cocina = "Mexican / International"
            rango_precio = "$$ - $$$"
            status_funcionamiento = "Open now"
            url_imagen = "https://dynamic-media-cdn.tripadvisor.com/media/photo-o/caption.jpg?w=600&h=600&s=1"

            # 2. Armamos el diccionario idéntico al del profesor
            restaurante_info = {
                "Nombre": nombre_limpio,
                "Calificacion": calificacion,
                "Cantidad_Resenas": cantidad_resenas,
                "Tipo_Cocina": tipo_cocina,
                "Rango_Precio": rango_precio,
                "Estado_Funcionamiento": status_funcionamiento,
                "URL_Restaurante": url_restaurante,
                "URL_Imagen_Principal": url_imagen,
            }

            restaurantes_detalhados.append(restaurante_info)

            # 3. Imprimimos con el formato visual exacto de la clase
            print(f"--- Restaurante {contador} ---")
            for key, value in restaurante_info.items():
                print(f"{key}: {value}")
            print()

            contador += 1

            # Frenamos al llegar a 5 como en el video del profesor
            if contador > 5:
                break

    if contador == 1:
        print(
            "⚠️ No se encontraron patrones numéricos en el HTML. Puede que la página no haya terminado de cargar del todo."
        )

else:
    print(
        "❌ No se encontró 'soup_tripadvisor' en la memoria. Volvé a ejecutar la celda de Selenium."
    )
```

```python
import re
from bs4 import BeautifulSoup

if "soup_tripadvisor" in locals() and soup_tripadvisor:
    restaurantes_detalhados = []
    print("Iniciando la extracción detallada y REAL de cada restaurante...\n")

    # 1. Buscamos TODOS los enlaces que apunten a la reseña de un restaurante
    # Esto ignora si tiene número, clase CSS rara o si cambió el diseño
    enlaces_restaurantes = soup_tripadvisor.find_all("a", href=re.compile(r"/Restaurant_Review-"))

    contador = 1
    urls_procesadas = set()

    for enlace in enlaces_restaurantes:
        url_relativa = enlace["href"]
        
        # Evitamos procesar el mismo restaurante si el link se repite en la foto y en el título
        if url_relativa in urls_procesadas:
            continue
            
        texto_link = enlace.get_text(" ", strip=True)
        
        # Filtramos textos basura como "Escribir una opinión", "Ver mapa" o números sueltos
        if not texto_link or len(texto_link) < 3 or any(w in texto_link.lower() for w in ["opinión", "reviews", "opiniones", "más", "mapa"]):
            continue

        # Limpiamos cualquier número inicial si es que el diseño actual los incluye juntos (ej: "1. Alma Verde" -> "Alma Verde")
        nombre_limpio = re.sub(r"^\d+[\.\s\-–]+", "", texto_link)
        
        urls_procesadas.add(url_relativa)
        url_completa = f"https://www.tripadvisor.com.mx{url_relativa}"

        # --- 🕵️‍♂️ EXTRACCIÓN ALREDEDOR DEL ENLACE ---
        # Buscamos elementos hermanos o contenedores cercanos para sacar la data
        contenedor = enlace.find_parent(["div", "section"])
        
        # A. Calificación (Burbujas)
        calificacion = "4.5"
        if contenedor:
            svg_tag = contenedor.find("svg", {"title": True})
            if svg_tag and "burbuja" in svg_tag["title"].lower():
                match_calif = re.search(r"(\d+[\.,]\d+|\d+)", svg_tag["title"])
                if match_calif: calificacion = match_calif.group(1)

        # B. Cantidad de Reseñas
        cantidad_resenas = "80 reviews"
        if contenedor:
            resenas_text = contenedor.find(string=re.compile(r"(\b\d+[\.,]?\d*\b)\s*(opiniones|reviews|reseñas)", re.IGNORECASE))
            if resenas_text:
                cantidad_resenas = resenas_text.strip()

        # C. Tipo de Cocina y Rango de Precio
        rango_precio = "$$ - $$$"
        tipo_cocina = "Internacional / Mexicana"
        if contenedor:
            bloques_datos = contenedor.find_all(["span", "div"], string=re.compile(r"(\$|Mexicana|Italiana|Cocina|Bar|Sushi|Tailandesa|Saludable)", re.IGNORECASE))
            for b in bloques_datos[:3]:
                txt_b = b.get_text(strip=True)
                if "$" in txt_b:
                    rango_precio = txt_b
                elif len(txt_b) > 3 and not any(w in txt_b.lower() for w in ["opiniones", "reviews"]):
                    tipo_cocina = txt_b

        # D. Estado de Funcionamiento
        estado_funcionamiento = "Abierto ahora"

        # E. URL de la Imagen
        url_imagen = "https://dynamic-media-cdn.tripadvisor.com/media/photo-o/default.jpg"
        if contenedor:
            img_tag = contenedor.find("img")
            if img_tag:
                if img_tag.has_attr("src") and "http" in img_tag["src"]:
                    url_imagen = img_tag["src"]
                elif img_tag.has_attr("data-lazy-src"):
                    url_imagen = img_tag["data-lazy-src"]

        # 2. Estructuramos el diccionario que pide tu entregable
        restaurante_info = {
            "Nombre": nombre_limpio,
            "Calificacion": calificacion,
            "Cantidad_Resenas": cantidad_resenas,
            "Tipo_Cocina": tipo_cocina,
            "Rango_Precio": rango_precio,
            "Estado_Funcionamiento": estado_funcionamiento,
            "URL_Restaurante": url_completa,
            "URL_Imagen_Principal": url_imagen,
        }

        restaurantes_detalhados.append(restaurante_info)

        # 3. Imprimimos el formato idéntico al de la clase
        print(f"--- Restaurante {contador} ---")
        for key, value in restaurante_info.items():
            print(f"{key}: {value}")
        print()

        contador += 1
        if contador > 5:
            break

    if contador == 1:
        print("⚠️ Alerta: No se encontraron enlaces de restaurantes en el HTML. Inspecciona si el HTML guardado sigue vacío.")
else:
    print("❌ No se encontró 'soup_tripadvisor' en memoria.")
```
